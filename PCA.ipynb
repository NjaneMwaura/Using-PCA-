```python
def correlation_matrix(df):
    plt.figure(figsize=(16, 12))
    cax = plt.imshow(df.corr(), cmap='jet', interpolation='nearest')
    plt.title('Wine dataset features correlation', fontsize=15)
    plt.xticks(range(len(df.columns)), df.columns, fontsize=9, rotation=90)
    plt.yticks(range(len(df.columns)), df.columns, fontsize=9)
    plt.colorbar(cax)
    plt.show()

correlation_matrix(df)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x=df.drop('Class', axis=1)
y=df['Class']
```

```python
x=scaler.fit_transform(x)
```

```python
dfx=pd.DataFrame(data=x, columns=df.columns[1:])
dfx.head()
```

```python
dfx.describe()
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=None)
```

```python
dfx_pca=pca.fit(dfx)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(range(1, len(dfx_pca.explained_variance_ratio_)+1), 
            dfx_pca.explained_variance_ratio_, s=200, alpha=0.75, c='orange', edgecolor='k')
plt.grid(True)
plt.title("Explained Variance Ratio of Principal Components", fontsize=25)
plt.xlabel("Principal Components", fontsize=15)
plt.ylabel("Explained Variance Ratio", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
dfx_trans = pca.transform(dfx)
```

```python
dfx_trans = pd.DataFrame(data=dfx_trans)
dfx_trans.head(10)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(dfx_trans[0],dfx_trans[1],c=df['Class'],edgecolors='k',alpha=0.75,s=150)
plt.grid(True)
plt.title("Class separation using first two principal components\n",fontsize=20)
plt.xlabel("Principal component-1",fontsize=15)
plt.ylabel("Principal component-2",fontsize=15)
plt.show()
```
